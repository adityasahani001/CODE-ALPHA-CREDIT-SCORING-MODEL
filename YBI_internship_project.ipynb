```python
# Import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read dataset
df = pd.read_csv('https://github.com/YBIFoundation/Pandas/raw/refs/heads/main/EcommerceEDA.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Missing value count
df.isna().sum()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'],errors='coerce')
```

```python
df.info()
```

```python
df['price'].median()
```

```python
# Missing value fill
df['price'].fillna(df['price'].median(), inplace = True)
```

```python
df.isna().sum()
```

```python
# Impute missing quantity with median
df['quantity'] = df['quantity'].fillna(df['quantity'].median())
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset= ['state','order_date'])
```

```python
df.info()
```

```python
# Question: Which product categories generate the highest total sales, and how do they compare?

df['category'].value_counts()
```

```python
df.groupby('category')['total_price'].sum().sort_values()
```

```python
df.columns
```

```python
y = df['payment_method']
X = df[['order_id', 'customer_id', 'order_date', 'category', 'product', 'price',
       'quantity', 'total_price', 'state', 'region']]
```

```python
import statsmodels.api as sm
print(sm.OLS(y,X).fit().summary())
```

```python
# Step: split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=2529)
```

```python
# Step: Model selection
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Step: Train
model.fit(X_train,y_train)
```

```python
# Step: Prediction
y_pred = model.predict(X_test)
```

```python
# Step: Evaluate
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```
